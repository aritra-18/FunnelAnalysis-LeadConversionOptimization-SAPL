```python
# Import pandas library
import pandas as pd
```

```python
# Read the "city_table"
df1= pd.read_csv("city_table.csv")
df1.head(2)
```

```python
# Read the "hostel_table"
df2= pd.read_csv("hostel_table.csv")
df2.head(2)
```

```python
# Rename the column
df2= df2.rename(columns={"id": "Hotel_reg_id"})
df2.head(2)
```

```python
# Read the "lead_visit_table"
df3= pd.read_csv("lead_visit_table.csv")
df3.head(2)
```

```python
# Rename column
df3= df3.rename(columns={"cityid": "city_id","id":"user_id"})
df3.head(2)
```

```python
# Read the "status_table"
df4= pd.read_csv("status_table.csv")
df4.head(2)
```

```python
# Merge "lead_visit_table" and "city_table"
df5= pd.merge(df3, df1,how= "left", on= "city_id")
df5.head(2)
```

```python
# Merge df5 and "hostel_table"
df6= pd.merge(df5, df2,how= "left", on= "hostel_id")
df6.head(2)
```

```python
# Rename column
df6= df6.rename(columns={"sustatus_id":"substatus_id"})
df6.head(2)
```

```python
# Merge df6 and "status_table"
merged_df= pd.merge(df6, df4,how= "left", on= "substatus_id")
merged_df.head(20)
```

```python
# Rename the columns
merged_df= merged_df.rename(columns={"city_id_x": "city_id", "city_id_y": "city_id_extra"})

# Drop the extra columns (caused by merging)
merged_df= merged_df.drop("city_id_extra", axis=1)
```

```python
# Check the tail of the dataset
merged_df.tail(20)
```

```python
# Save as csv
merged_df.to_csv("merged_data.csv", index= False)
```
